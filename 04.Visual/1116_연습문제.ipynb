```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# 한글폰트 사용
mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
# 1번
df = pd.read_csv('samples/cafe_2year.csv')
df.head()
```

```python
a = df.set_index('기준월')
b = a[2::3]
x = b.index.astype(str)
y = b['카페수'].astype(str)
ylim=(min(y), max(y))
plt.title('전국의 월별 카페수', color='green', fontsize=16)
plt.xlabel('년월')
plt.ylabel('카페수')
plt.bar(x,y)
```

```python
# 2번
o_doctor = pd.read_csv('samples/doctor_2019.csv')
z = o_doctor[['일반의 수', '인턴 수', '레지던트 수', '전문의 수']].groupby(o_doctor['지역']).sum()
c = z.loc[['서울', '부산', '대구', '인천', '대전', '광주', '울산']]
c
```

```python
# 4번
doc = pd.read_csv('samples/doctor22.csv')
doc
```

```python
x = doc['지역']
y = doc['전문의 수']
plt.title('서울 및 광역시 종합병원 전문의 수', color='green', fontsize=16)
plt.scatter(x, y)
plt.xlabel('시')
plt.ylabel('의사수')
plt.show()
```

```python
# 5번
x = doc['지역']
y1 = doc['일반의 수']
y2 = doc['인턴 수']
y3 = doc['레지던트 수']
y4 = doc['전문의 수']

plt.plot(x, y1, 'ro-', label='일반의')
plt.plot(x, y2, 'b*--', label='인턴')
plt.plot(x, y3, 'ys:', label='레지던트')
plt.plot(x, y4, 'mp-.', label='전문의')
plt.grid(True, axis='y', color='gray', alpha=0.5, linestyle='--')
plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
plt.xlabel('시')
plt.ylabel('의사수')
plt.legend()
plt.title('서울 및 광역시 종합병원 전문의 수', color='green', fontsize=16)

plt.show()
```

```python
# 6번
doc2 = doc.set_index('지역')

plt.axis('equal')
doc_pct = doc2.T['서울']
labels=['일반의', '인턴', '레지던트', '전문의']
explode = [0, 0.1, 0, 0]
doc_pct.plot.pie(autopct='%.1f%%', labels=labels, startangle=90, explode=explode)
plt.title('서울 종합병원 의사 유형별 분포', color='green', fontsize=16)
plt.show()
```
