```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 한글폰트 사용
mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
#내가한것 (수정필요)
x = np.arange(0, 2*np.pi, 0.01)
si = np.sin(x)
co = np.cos(x)

plt.plot(x, si)
plt.plot(x, co)
plt.title("Sin, Cos 그래프", size=16)
plt.xticks([0, 1, 2, 3, 4, 5], ['0', 'pi/2', 'pi', '3pi/2', '2pi'])
plt.yticks([-1, 0, 1], ['LOW', "ZERO", "HIGH"])
plt.grid(True)

plt.show()
```

```python
# 강사님이 한 것
X = np.linspace(0, 2*np.pi, 256)
S = np.sin(X)
C = np.cos(X)
plt.title("Sin, Cos 그래프")
plt.plot(X, S, label="sine")
plt.plot(X, C, label="cosine")
plt.grid(True)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$-\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$']) # LaTeX
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.legend(loc='best')

plt.show()
```
