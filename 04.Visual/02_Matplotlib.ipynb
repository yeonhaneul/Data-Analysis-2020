#### 여러 개의 그래프 그리기

### Matplotlib 그래프 영역 채우기

```python
# (x[1], y[1]), (x[2], y[2]), (x[1], 0), (x[2], 0)를 잇는 영역

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[1:3], y[1:3], alpha=0.5) ## fill_between() 사용, alpha는 투명도
# plt.fill_betweenx(y[2:4], x[2:4], color='pink', alpha=0.5)  ## fill_betweenx() 사용
plt.grid(True)
plt.axhline(0); plt.axvline(0)

plt.show()
```

```python
# (x[2], y[2]), (x[3], y[3]), (0, y[2]), (0, y[3])를 잇는 영역
# between과는 다르게 x가 0이된다.

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_betweenx(y[2:4], x[2:4], color='pink', alpha=0.5)  ## fill_betweenx() 사용
plt.grid(True)
plt.axhline(0); plt.axvline(0)

plt.show()
```

```python
# 두 그래프 사이의 영역 채우기
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [0, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.grid(True)
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='lightgray', alpha=0.5)

plt.show()
```

```python
# 두 그래프 사이의 영역 채우기
x = np.arange(1.0, 4.0, 0.01)
y1 = x ** 2
y2 = x ** 2 / 2

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label'); plt.ylabel('Y-Label')
plt.title('두 그래프 사이의 영역 채우기')
plt.grid(True)
plt.fill_between(x[100:200], y1[100:200], y2[100:200], color='lightgray', alpha=0.5)

plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label'); plt.ylabel('Y-Label')
plt.title('임의의 영역 채우기')
plt.grid(True)
plt.fill([1.9, 1.9, 2.5, 3.1, 3.1], [2, 5, 9, 11, 8], color='lightgray', alpha=0.5)

plt.show()
```

### Matplotlib 수직선/수평선 표시하기

```python
# 기본 사용 - axhline/axvline

a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.axhline(1, 0, 0.55, color='gray', linestyle='--', linewidth='1')
plt.axvline(1, 0, 0.16, color='lightgray', linestyle=':', linewidth='2')

plt.axhline(5.83, 0, 0.95, color='gray', linestyle='--', linewidth='1')
plt.axvline(1.8, 0, 0.95, color='lightgray', linestyle=':', linewidth='2')

plt.show()
```

```python
# 기본 사용 - hlines/vlines

a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.hlines(4, 1, 1.6, colors='pink', linewidth=3)
plt.vlines(1, 1, 4, colors='pink', linewidth=3)

plt.show()
```

### Matplotlib 막대 그래프 그리기

```python
x = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)
plt.show()
```

```python
# 막대그래프 스타일 꾸미기
x = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.bar(x, values, width=0.6, align='edge', color="springgreen",
        edgecolor="gray", linewidth=3, tick_label=years, log=True)
plt.title('스타일 꾸미기')
plt.show()
```

### Matplotlib 수평 막대 그래프 그리기

```python
# barh(y, values)만 줘도 그래프는 정상적으로 나온다.
y = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.barh(y, values, height=-0.6, align='edge', color="springgreen",
        edgecolor="gray", linewidth=3, tick_label=years, log=False)
plt.show()
```

### Matplotlib 산점도 그리기

```python
# 산점도 기본
np.random.seed(2020)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x, y)
plt.show()
```

```python
# jitter - 같은 위치에 있는 점이 조금 옆으로 나와서 겹쳐보이지 않게해줌
np.random.seed(2020)    # Reproducible random state

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

### Matplotlib 히스토그램 그리기
### (히스토그램 : 도수분포표를 그래프로 나타낸 것)

```python
# hist
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71, 80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight)
plt.show()
```

```python
np.random.seed(2020)
x = np.random.randn(1000)
plt.hist(x)
plt.show()
```

```python
# 위의 히스토그램을 조금 더  세분화
np.random.seed(2020)
x = np.random.randn(100000)
plt.hist(x, bins=100)
plt.show()
```

```python
x = np.linspace(-4.0, 4.0, 801)
nd = np.exp(-x*x/2) / np.sqrt(2*np.pi)
plt.plot(x, nd)
plt.axhline(0, alpha=0.5); plt.axvline(0, alpha=0.5)
plt.show()
```

```python
# 범주차이로, 선형 히스토그램이 잘 표시되지 않는다.
np.random.seed(2020)
hx = np.random.randn(100000)
plt.hist(hx, bins=100)

x = np.linspace(-4.0, 4.0, 801)
nd = np.exp(-x*x/2) / np.sqrt(2*np.pi)
plt.plot(x, nd, color="red", linewidth=3)
plt.axhline(0, alpha=0.5); plt.axvline(0, alpha=0.5)

plt.show()
```

```python
#여러개의 히스토그램 그리기
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step', label='a')
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled', label='b')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step', label='c')
plt.legend()
plt.show()
```

### Matplotlib 파이 차트 그리기

```python
# 반시계방향으로 생성된다.
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
# startangle을 지정해주어, 원하는 각도부터 시작을 할 수 있다.
# counterclock = True라면 반시계방향, False라면 시계방향으로 돌아간다.
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False)
plt.show()
```

```python
# 중심에서 벗어나는 파이
# explode를 통하여 중심에서 벗어날 수 있는 각도를 정해줄 수 있다.
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.1]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```
