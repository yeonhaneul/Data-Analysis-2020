#### Axes 객체와 subplot 명령

```python
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1) #cos2𝝿*e(-x) , 점점 진폭이 줄어드는 그래프
y2 = np.cos(2 * np.pi * x2) #cox2𝝿x

ax1 = plt.subplot(2, 1, 1) #(2개의 행, 1개의 열로 된 그래프의 첫번째 그래프)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2) #(2개의 행, 1개의 열로 된 그래프의 두번째 그래프)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

#plt.tight_layout()
plt.show()
```

```python
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1) #cos2𝝿*e(-x) , 점점 진폭이 줄어드는 그래프
y2 = np.cos(2 * np.pi * x2) #cox2𝝿x

ax1 = plt.subplot(2, 1, 1) #(2개의 행, 1개의 열로 된 그래프의 첫번째 그래프)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2) #(2개의 행, 1개의 열로 된 그래프의 두번째 그래프)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout() # 자동으로 명시된 여백에 관한 파라미터를 조절해준다.(pad=1.07, h_pad=None, w_pad=None, rect=None)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
```

```python
#위의 표를 반복문을 사용해서 그리기

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
# sharex를하면 1,3 / 2,4번째 그래프의 x축이 공유된다.
# sharey를 하면 1,2 / 3,4번째 그래프의 y축이 공유된다.

np.random.seed(0)
for i in range(2):
    for k in range(2):
        ax = axes[i, k]
        ax.plot(np.random.rand(5))
        ax.set_title(f"axes {2*i+k+1}")

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0.0, 5.0, 100)

fig, axes = plt.subplots(2, 1, sharex=True)

for i in range(2):
    ax=axes[i]
    y = np.cos(2*np.pi*x1) * np.exp(-x1) if i == 0 else np.cos(2*np.pi*x2)
    ylabel = 'Damped oscillation' if i == 0 else 'Undamped'
    ax.plot(x, y)
    ax.axhline(0)
    ax.set_ylim(-1,1)
    ax.set_ylabel(ylabel)
    #ax.set_xlabel('time (s)')

plt.xlabel('time (s)')
fig.suptitle('A tale of 2 subplots')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 8), sharey=True)
#list에 대하여 enumerate를 하면, 인덱스와 요소가 각각 나온다.
for i, spec in enumerate(['setosa', 'versicolor', 'virginica']):
    x = iris[iris.species == spec].sepal_length
    y = iris[iris.species == spec].sepal_width
    z = iris[iris.species == spec].petal_length
    k = iris[iris.species == spec].petal_width
    ax = axes[i]
    ax.boxplot([x, y, z, k], labels=["sepal length", "sepal width", "petal length", "petal width"])
    ax.set_title(f'{spec} 품종')
plt.show()
```

#### Anscombe's Quartet

```python
# ppt 13.Matplotlib.pdf의 9page
ans = sns.load_dataset('anscombe')
ans.shape
```

```python
ans.head()
```

```python
ans.groupby('dataset').describe()
```

```python
ds1 = ans[ans['dataset']=='I']
ds2 = ans[ans['dataset']=='II']
ds3 = ans[ans['dataset']=='III']
ds4 = ans[ans['dataset']=='IV']
```

```python
fig = plt.figure(figsize=(10, 8))
ax1 = plt.subplot(2, 2, 1)
ax2 = plt.subplot(2, 2, 2)
ax3 = plt.subplot(2, 2, 3)
ax4 = plt.subplot(2, 2, 4)

ax1.plot(ds1['x'], ds1['y'], 'o')
ax2.plot(ds2['x'], ds2['y'], 'o')
ax3.plot(ds3['x'], ds3['y'], 'o')
ax4.plot(ds4['x'], ds4['y'], 'o')

ax1.set_title('Dataset I')
ax2.set_title('Dataset II')
ax3.set_title('Dataset III')
ax4.set_title('Dataset IV')

fig.suptitle("Anscombe's Quartet")
fig.tight_layout()
plt.show()
```

```python
# 위의 표를 for문을 사용하여 만들기
fig, axes = plt.subplots(2, 2, figsize=(12,10), sharex=True, sharey=True)

for i, num in enumerate(['I', 'II', 'III', 'IV']):
    ds = ans[ans['dataset'] == num]
    x = ds.x; y = ds.y
    ax = axes[i//2][i%2]
    ax.plot(x, y, 'o', markersize=10)
    ax.plot([3, 20], [4.5, 13], 'r-', lw=2)
    ax.set_title(f'Dataset {num}')
    
fig.suptitle("Anscombe's Quartet")
fig.tight_layout()
fig.subplots_adjust(top=0.88) # title 위치조정
plt.show()
```

### Axis 객체와 축

```python
# 2개의 다른 y축 사용하기

fig, ax0 = plt.subplots()
ax1 = ax0.twinx()

ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)

ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)

ax0.set_xlabel("공유되는 x축")

fig.legend(loc='upper right', bbox_to_anchor=(0.86, 0.88))
plt.show()
```
