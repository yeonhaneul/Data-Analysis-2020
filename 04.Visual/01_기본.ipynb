# 데이터 시각화

### 시각화 패키지 맷플롯리브 소개

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

### 라인 플롯

```python
plt.plot([1, 4, 9, 16]) # y좌표
plt.title('Line Plot')
plt.show()
# x좌표는 기본적으로 y좌표의 갯수만큼(4개라서 0, 1, 2, 3)이 생성된다.
# 그래프를 더블클릭 후 왼쪽 상단의 save를 누르면 그래프를 저장 할 수 잇다.
```

```python
plt.plot([1, 4, 9, 16])
plt.title('Line Plot')
plt.grid(True) # 표에 그리드가 보이게
plt.show()
```

```python
plt.plot([1, 4, 9, 16])
plt.title('Line Plot')
plt.grid(True) # 표에 그리드가 보이게
plt.xlabel('X Axis') # x축의 title(x 레이블)
plt.ylabel('Y Axis') # y축의 title(y 레이블)
plt.show()
```

```python
plt.title("x ticks")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16]) #plt.plot(x좌표, y좌표)
plt.grid(True)
plt.show()
```

```python
plt.title("x 틱스") # 한글은 그냥 사용하면 깨진다.
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.grid(True)
plt.show()
```

```python
# 한글을 사용 할 수 있도록
import matplotlib.font_manager

matplotlib.font_manager._rebuild()
sorted([f.name for f in matplotlib.font_manager.fontManager.ttflist if f.name.startswith("Malgun")])
```

```python
# 폰트 설정
mpl.rc('font', family='Malgun Gothic')
# 유니코드에서  음수 부호설정
mpl.rc('axes', unicode_minus=False)
```

```python
plt.title('한글 제목')
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.xlabel("엑스축 라벨")
plt.ylabel("와이축 라벨")
plt.show()
```

```python
# 참고용 (폰트 설정하는법)
font1 = {'family': 'NanumMyeongjo', 'size': 24, 
         'color':  'black'}
font2 = {'family': 'NanumBarunpen', 'size': 18, 'weight': 'bold',
         'color':  'darkred'}
font3 = {'family': 'NanumBarunGothic', 'size': 12, 'weight': 'light',
         'color':  'blue'}

plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.title('한글 제목', fontdict=font1)
plt.xlabel('엑스 축', fontdict=font2)
plt.ylabel('와이 축', fontdict=font3)
plt.show()
```
