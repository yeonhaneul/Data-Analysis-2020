### address3.csv를 읽어서 지도위에 표시하기

```python
df = pd.read_csv('address3.csv')
df
```

```python
# 마커 아이콘 - 데이터프레임에 추가
df['icon'] = ['map-marker', 'cloud', 'gift', 'info-sign', 'ok-circle']
df
```

```python
# 날씨에 따른 마커의 색상
color_dict = {'Clear': 'blue', 'Snow': 'white', 'Rain': 'gray', 'Extreme':'red', 'Clouds': 'orange', 'Mist': 'green'}
```

```python
mapping = folium.Map(location=[df.lat.mean(), df.lng.mean()], zoom_start=12)
mapping
```

```python
for i in df.index:
    folium.Marker(
        location=[df.lat[i], df.lng[i]],
        popup=df.bldg[i], # 아이콘을 누르면 나타난다
        tooltip=f'{df.desc[i]}, {df.temp[i]}', # 아이콘에 마우스를 갖고가면 나타난다
        icon=folium.Icon(color=color_dict[df.weather[i]], icon=df.icon[i])
    ).add_to(mapping)

mapping
```

```python
folium.CircleMarker(
        location=[37.522277, 126.805003],
        radius=50,
        popup='원 그림',
        color='#3186cc',
        fill=True,
        fill_color='#3186cc'
    ).add_to(mapping)
mapping
```
