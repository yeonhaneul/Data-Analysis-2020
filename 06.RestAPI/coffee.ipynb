```python
import pandas as pd
import folium
```

```python
df = pd.read_csv('상권정보서울.csv', encoding='utf8', sep='|')
df
```

```python
df2 = df.loc[:, ['상호명', '지점명', '상권업종중분류명', '시도명', '시군구명', '도로명주소', '경도', '위도']]
df2.head()
```

```python
df3 = df2[df2['상권업종중분류명'] == '커피점/카페']
df4 = df3[df3['시도명'] == '서울특별시']
df5 = df4[df4['시군구명'] == '강서구']
df5.tail()
```

```python
df6 =df5[df5['상호명'].str.contains('스타벅스|투썸|이디|빽')]
df6['지점명'] = df6['지점명'].fillna(' ')
df6.tail()
```

```python
mapping = folium.Map(location=[df6.위도.mean(), df6.경도.mean()], zoom_start=13)
for i in df6.index:

    folium.Marker(
        location=[df6.위도[i], df6.경도[i]],
        popup=df6.도로명주소[i],
        tooltip=f'{df6.상호명[i]}{df6.지점명[i]}',
    ).add_to(mapping)

mapping
```

```python
star = df6[df6['상호명'].str.contains('스타벅스|STAR')]
ed = df6[df6['상호명'].str.contains('이디')]
baek = df6[df6['상호명'].str.contains('빽')]
two = df6[df6['상호명'].str.contains('투썸')]
```

```python
mapping2 = folium.Map(location=[df6.위도.mean(), df6.경도.mean()], zoom_start=14)

for s in star.index:
    folium.Marker(
        location=[star.위도[s], star.경도[s]],
        popup=star.도로명주소[s],
        tooltip=f'{star.상호명[s]}{star.지점명[s]}',
        icon=folium.Icon(color='green', icon='star')
    ).add_to(mapping2)

for e in ed.index:
    folium.Marker(
        location=[ed.위도[e], ed.경도[e]],
        popup=ed.도로명주소[e],
        tooltip=f'{ed.상호명[e]}{ed.지점명[e]}',
        icon=folium.Icon(color='blue', icon='euro')
    ).add_to(mapping2)

for b in baek.index:
    folium.Marker(
        location=[baek.위도[b], baek.경도[b]],
        popup=baek.도로명주소[b],
        tooltip=f'{baek.상호명[b]}{baek.지점명[b]}',
        icon=folium.Icon(color='purple', icon='bold')
    ).add_to(mapping2)

for t in two.index:
    folium.Marker(
        location=[two.위도[t], two.경도[t]],
        popup=two.도로명주소[t],
        tooltip=f'{two.상호명[t]}{two.지점명[t]}',
        icon=folium.Icon(color='red', icon='text-width')
    ).add_to(mapping2)

mapping2
```

```python
star['a'] = '스타벅스'
ed['a'] = '이디야'
baek['a'] = '빽다방'
two['a'] = '투썸'
```

```python
cafe = pd.concat([star, ed, baek, two])
cafe.count()
```

```python
icon_list = []

for i in cafe.index:
    if cafe.a[i] == '스타벅스':
        icon_list.append('star')
    elif cafe.a[i] == '이디야':
        icon_list.append('euro')
    elif cafe.a[i] == '빽다방':
        icon_list.append('bold')
    elif cafe.a[i] == '투썸':
        icon_list.append('text-width')
```

```python
cafe['icon'] = icon_list
cafe.head()
```

```python
color_dict = {'스타벅스':'green', '이디야':'blue','빽다방':'purple','투썸':'red'}
```

```python
mapping3 = folium.Map(location=[cafe.위도.mean(), cafe.경도.mean()], zoom_start=14)

for i in cafe.index:
    folium.Marker(
        location=[cafe.위도[i], cafe.경도[i]],
        popup=cafe.도로명주소[i],
        tooltip=f'{cafe.상호명[i]}{cafe.지점명[i]}',
        icon=folium.Icon(color=color_dict[cafe.a[i]], icon=cafe.icon[i])
    ).add_to(mapping3)

del cafe['a']
mapping3
```
